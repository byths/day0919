## 1. 판다스 시작- 파일을 DataFrame으로 로딩, 기본 API
## 판다스의 핵심 객체: DataFrame

```python
#pandas를 pd로 에일리어스(alias)해 임포트
import pandas as pd

#호출 시 파일명 인자로 들어온 파일을 로 딩해 DataFrame객체로 반환
titanic_df=pd.read_csv('titanic_train.csv')
titanic_df.head(3)

print('titanic 변수 type:', type(titanic_df))
titanic_df
```

```python
#디폴트는 5, 맨 앞 3개의 로우만 표출
titanic_df.head(3)

#shape변수는 DataFrame의 행과 열을 튜플 형태로 반환
print('DataFrame 크기: ', titanic_df.shape)

#칼럼의 타입, NULL 데이터 개수, 데이터 분포 등 조회 가능
#info()메서드를 통해 총 데이터 건수, 데이터 타입, NULL건수를 알 수 있음
titanic_df.info()

#describe() 메서트를 통해 칼럼별 숫자형 데이터값의 n-percentile분포도, 평균값, 최댓값, 최솟값 나타냄. 오직 숫자형칼럼의 분포도만 조사
titanic_df.describe()
```

```python
#DataFrame의 []연산자 내부에 칼럼명을 입력하면 Series 형태로 반환
#value_counts 메서드: 해당 칼럼값의 유형과 건수 확인가능, 시리즈객체에서만 사용가능
value_counts=titanic_df['Pclass'].value_counts()
print(value_counts)

```

Pclass 값 3이 491개, 1이 216개, 2가 184개 임을 의미. 많은 건수 순서로 정렬되어 값을 반환.
또한 이 결과 자체가 시리즈 객체. 앞의 3,1,2는 결과 시리즈의 index. 

## 2. DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환
## 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기
### DataFrame은 2차원 데이터 이므로 2차원 이하의 데이터들만 DataFrame으로 변환될 수 있음!

```python
import numpy as np

#먼저 1차원 데이터
col_name1=['col1']
list1=[1,2,3]
array1=np.array(list1)
print('array1 shape:', array1.shape)

#리스트를 이용해 DataFrame 생성
df_list1=pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame: \n', df_list1)

#넘파이 ndarray를 이용해 DataFrame 생성
df_array1=pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame: \n', df_array1)
```

```python
#2차원 데이터
#3개의 칼럼명 필요함
col_name2=['col1','col2','col3']

#2행 3열의 리스트와 ndarray생성 후 이를 DataFrame으로 변환
list2=[[1,2,3],[11,12,13]]
array2=np.array(list2)
print('array2 shape:', array2.shape)
df_list2=pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame: \n', df_list2)
df_array2=pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame: \n', df_array2)
```

```python
#딕셔너리는 DataFrame으로 변환
#Key는 문자열 칼럼명으로 매핑, Value는 리스트형(또는 ndarray형) 칼럼 데이터로 매핑
dict={'col1':[1,11],'col2':[2,22],'col3':[3,33]}
df_dict=pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame: \n', df_dict)
```

## DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기
### 많은 머신러닝 패키지가 기본 데이터 형으로 넘파이 ndarray를 사용. 따라서 DataFrame을 다시 넘파이 ndarray로 변환하는 경우가 빈번하게 발생함.
### values를 이용한 ndarray로의 변환은 반드시 기억!!

```python
#DataFrame을 ndarray로 변환
array3=df_dict.values
print('df_dict.values 타입:', type(array3),'df_dict.values shape:', array3.shape)
print(array3)
```

```python
#DataFrame을 리스트로 변환
list3=df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

#DataFrame을 딕셔너리로 변환
dict3=df_dict.to_dict('list')
print('\n df_dict.to dict() 타입:', type(dict3))
print(dict3)
#to_dict('list')인자로 list를 입력하면 딕셔너리의 값이 리스트형으로 반환
```

## 3. DataFrame의 칼럼 데이터 세트 생성과 수정

```python
#타이타닉 데이터 프레임에 새로운 칼럼을 추가하고 일괄적으로 0 할당
titanic_df['Age_0']=0
titanic_df.head(3)
```

```python
#기존 칼럼 Series의 데이터를 이용해 새로운 칼럼 Series 만들기
titanic_df['Age_by_10']=titanic_df['Age']*10
titanic_df['Family_No']=titanic_df['SibSp']+titanic_df['Parch']+1
titanic_df.head(3)
```

```python
#DataFrame 내의 기존 칼럼 값 업데이트
titanic_df['Age_by_10']=titanic_df['Age_by_10']+100
titanic_df.head(3)
```

## 4. DataFrame 데이터 삭제
## drop() 메서드원형: DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
## labels, axis, inplace

```python
#axis 0은 로우 방향 축, axis 1은 칼럼 방향 축
#drop() 메서드에 axis=1을 입력하면 칼럼 축 방향으로 드롭 수행
titanic_drop_df=titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
```

```python
#inplace=True로 설정하면 원래 본인의 DataFrame의 데이터를 삭제
#여러 개의 컬럼을 삭제할 때 리스트 형태로 labels 파라미터 입력
drop_result=titanic_df.drop(['Age_0','Age_by_10','Family_No'],axis=1,inplace=True)
print('inplace=True 로 drop 후 반환된 값:', drop_result)
titanic_df.head(3)
```

반환된 값이 None. inplace를 트루로 설정한 채로 반환 값을 다시 자신의 데이터프레임 객체로 할당하지 말라는데 그럼 어떻게 ?? 모르겠음

```python
#axis=0으로 설정해 index 로우 삭제
pd.set_option('display.width',100)
pd.set_option('display.max_colwidth',15)
print('#### before axis 0 drop ####')
print(titanic_df.head(3))
#index 0,1,2에 위치한 로우를 삭제
titanic_df.drop([0,1,2], axis=0, inplace=True)

print('#### after axis 0 drop ####')
print(titanic_df.head(3))
```

인덱스 삭제 후 맨 앞 3개 데이터의 인덱스는 3,4,5이다. 
drop() 메서드에서 axis와 inplace인자를 적용해 dataframe을 변경하는 방식 정리
* axis: 로우를 삭제할 때는 axis=0, 칼럼을 삭제할 때는 axis=1로 설정
* 원본 DataFrame은 유지하고 드롭된 DataFrame을 새롭게 객체 변수로 받고 싶다면 inplace=False로 설정
* 원본 데이터 프레임에 드롭된 결과를 적용할 경우에는 inplace=True를 적용
* 원본 데이터프레임에서 드롭된 데이터프레임을 다시 원본 데이터프레임 객체 변수로 할당하면 원본 데이터프레임에서 드롭된 결과를 적용할 경우와 같음

## 5. Index 객체

```python
#위에 drop하고 했으니 원본 파일 다시 로딩
titanic_df=pd.read_csv('titanic_train.csv')

#인덱스 객체 추출
indexes=titanic_df.index

#인덱스 객체를 실제 값 array로 변환, 인덱스 객체의 values 속성으로 ndarray값 알 수 있음
print('Index 객체 array값: \n', indexes.values)
```

```python
#인덱스 객체는 식벽성 데이터를 1차원 array로 갖고 있음. 또한 ndarray와 유사하게 단일 값 반환 및 슬라이싱도 가능
print(type(indexes.values))
print(indexes.values.shape)
print(indexes[:5].values)
print(indexes.values[:5])
print(indexes[6])
```

### 한번 만들어진 DataFrame 및 Series의 Index 객체는 함부로 변경할 수 없으며 오직 식별용으로만 사용되어 연산에서도 제외됨.

```python
indexes[0]=5
```

```python
series_fair=titanic_df['Fare']
print('Fair Sereis max 값:', series_fair.max())
print('Fair Sereis sum 값:', series_fair.sum())
print('sum() Fair Sereis:', sum(series_fair))
print('Fair Sereis +3:\n', (series_fair+3).head(3))
```

```python
#reset_index() 메서드를 수행하면 새롭게 인덱스를 할당하고 기존 인덱스는 데이터 프레임에 'index'라는 새로운 칼럼으로 추가됨
titanic_reset_df=titanic_df.reset_index(inplace=False)
titanic_reset_df.head(3)
```

reset_index()는 인덱스가 연속된 int 숫자형이 아닐 경우에 주로 사용. 또한 시리즈 객체에 사용할 경우 데이터 프레임이 반환됨.

```python
print('### before reset_index ###')
value_counts=titanic_df['Pclass'].value_counts()
print(value_counts)
print('value_counts 객체 변수 타입:', type(value_counts))
new_value_counts=value_counts.reset_index(inplace=False)
print('### After reset_index ###')
print(new_value_counts)
print('new_value_counts 객체 변수 타입:', type(new_value_counts))
```
